```python
import numpy as np
from GA_Clustering import GACluster
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from sklearn.metrics import silhouette_score
```

```python
n_samples = 1500
random_state = 170
X_org, y = make_blobs(n_samples=n_samples, random_state=random_state)
```

```python
X = np.vstack((X_org[y == 0][:500], X_org[y == 1][:100], X_org[y == 2][:10]))

plt.figure(figsize=(8, 8))
plt.scatter(X[:, 0], X[:, 1], c=['b'])
plt.show()
```

### KMeans

```python
model = KMeans(n_clusters=3)
```

```python
model.fit(X)
```

```python
plt.figure(figsize=(8, 8))
plt.scatter(X[:, 0], X[:, 1], c= model.labels_)
plt.title('KMeans inertia: {} silhouette_score: {}'.format(round(model.inertia_), round(silhouette_score(X, model.labels_),2)))
plt.show()
```

### GACluster

```python
model = GACluster(n_clusters=3, max_iter=50, population_size=20, elitism_size=2,
                  category='tournament', tournament_size=5)
```

```python
model.fit(X)
```

```python
silhouette_score(X, model.labels_)
```

```python
model = GACluster(n_clusters=3, max_iter=20, population_size=10, elitism_size=2)
```

```python
model.fit(X)
```

```python
silhouette_score(X, model.labels_)
```
